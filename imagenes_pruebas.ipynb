```python
def decode_lsb(image_path):
    try:
        image = Image.open(image_path)
        data = ""
        width, height = image.size

        for y in range(height):
            for x in range(width):
                pixel = list(image.getpixel((x, y)))
                binary_data = ""
                for color in pixel:
                    binary_data += format(color, '08b')[-1]  # Obtiene el bit menos significativo

                data += binary_data

        # Divide los datos binarios en bytes
        data_bytes = [data[i:i+8] for i in range(0, len(data), 8)]
        message = ""
        for byte in data_bytes:
            message += chr(int(byte, 2))

        return message
    except Exception as e:
        print("Error al decodificar la imagen:", str(e))
        return None

# Ejemplo de uso:
image_path = 'ejemplo01.jpg'
decoded_message = decode_lsb(image_path)
if decoded_message:
    print("Mensaje decodificado:", decoded_message)
```

```python
#Ejercicio 59
def unionLista(l1,l2,l3):
    solution=[]
    listas=[l1,l2,l3]
    n=0
    while n <=len(listas)-1:
        for i in listas[n]:
            if i not in solution:
                solution.append(i)
        n+=1
    return sorted(solution)
print(unionLista([4,2,3],[],[4,-7,9]))
```

```python
#Ejercicio 57
def invertirFrase(frase):
    return ' '.join(frase.split(' ')[::-1])
print(invertirFrase('Hola a todo el mundo'))

```

```python
#ejercicio 56
def filtrarPalabras(frase,longitudMinima):
    return' '.join([x for x in frase.split(' ') if len(x)>=longitudMinima])
    
print(filtrarPalabras('Hola a todos',4))
```

```python
#Ejercicio 55
def posicionEltLista(L,x):
    solution=[]
    for i in range (len(L)):
        if L[i]==x:
            solution.append(i) 
    return solution
    
print(posicionEltLista([1,2,3,6,8,7,3],3))
```
